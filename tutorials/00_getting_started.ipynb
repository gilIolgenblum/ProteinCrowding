# Getting Started with ProteinCrowding

Welcome to the `ProteinCrowding` (via `fh_crowding`) tutorials! This package implements the Flory-Huggins based thermodynamic model for protein stability in crowded environments.

## What this package does
The package allows you to:
1. Simulate the effect of cosolutes (crowders) on protein stability ($\Delta\Delta G$, $\Delta\Delta H$, $T\Delta\Delta S$).
2. Fit experimental stability data to extract soft interaction parameters ($\varepsilon$, $\varepsilon_{TS}$) for both binary (one cosolute) and ternary (two cosolutes) systems.

## Scientific background
The physical framework relies on:
- **Excluded Volume ($\nu$):** Hard-core repulsions that always stabilize the compact/folded state.
- **Flory-Huggins Interactions ($\chi$):** Nonideal solvent-cosolute interactions.
- **Soft Interactions ($\varepsilon$):** Weak attractive/repulsive interactions between the protein and the cosolute.

## Available Workflows
We provide several classes for modeling:
- `fh_crowding.BinaryCrowdingModel`: For single cosolute systems.
- `fh_crowding.TernaryCrowdingModel`: For systems with two distinct cosolutes.
- `fh_crowding.Protein`: Defines the protein's Solvent Accessible Surface Area (SASA).
- `fh_crowding.Cosolute` and `fh_crowding.CosoluteMixture`: Define the properties of the crowders.

## Units and Notation
Consistent units are critical for correct thermodynamic modeling.
- **Concentration:** The core model operates using **volume fraction ($\phi$)**. You must supply concentrations in $\phi$, or if you supply molarity, ensure you convert it.
- **Energy:** The package expects experimental energies (like $\Delta\Delta G$) to be in **kJ/mol**. If your data is in kcal/mol, convert it by multiplying by $4.184$.
- **Symbols used in the code:**
  - `dG` -> $\Delta\Delta G$
  - `dH` -> $\Delta\Delta H$
  - `TdS` -> $T\Delta\Delta S$
  - `eps` -> $\varepsilon$
  - `epsTS` -> $\varepsilon_{TS}$
  - `chi` -> $\chi$
  - `nu` -> $\nu$
